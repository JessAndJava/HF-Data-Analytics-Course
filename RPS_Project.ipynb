```python
def get_move(player):
    # Ask for the move and clean up any accidental capital letters or spaces to avoid errors
    move = input(player + ", choose rock, paper, or scissors: ").lower().strip()

    # Customized message if player doesn't choose between rock, paper or scissors: This keeps looping until they type a real move
    while move != "rock" and move != "paper" and move != "scissors":
        print("The Vulcan mind trick will not work here! Choose rock, paper or scissors.")
        move = input(player + ", choose rock, paper, or scissors: ").lower().strip()

    return move

def decide_winner(p1, p2):
    # 1. Check for a tie first
    if p1 == p2:
        return "tie"

    # 2. Check every way Player 1 can win
    if p1 == "rock" and p2 == "scissors":
        return "player1"
    if p1 == "scissors" and p2 == "paper":
        return "player1"
    if p1 == "paper" and p2 == "rock":
        return "player1"

    # 3. If it wasn't a tie or a Player 1 win, Player 2 wins by default
    return "player2"


# ----RPS Game... Here we go!----

# My greeting: Prints once at the very start
print("Greetings, Professor Imran.")

play_again = "yes"

while play_again == "yes":
    # Get Player 1's move
    player1 = get_move("Player 1")

    # Hide the move so Player 2 can't see it (20 spaces)
    print("\n" * 20)

    # Get Player 2's move
    player2 = get_move("Player 2")

    print("Player 1 chose:", player1)
    print("Player 2 chose:", player2)

    # Consider this the 'referee' which determines the winner
    winner = decide_winner(player1, player2)

    # Customized messages for a win or a tie
    if winner == "tie":
        print("What a strange game. The only winning move is not to play.")
    elif winner == "player1":
        print("Player 1 wins. Congratulations!")
    else:
        print("Player 2 wins. Congratulations!")

    # Play again prompt
    play_again = input("Shall we play a new game? ").lower().strip()

# If anything but yes is answered, then the game stops/over
print("Game over, Professor Imran.")
```
