```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df.head() #displays the first 5 rows by default
```

```python
df.shape #seeing how big the dataset is (rows and columns)
```

```python
df.columns = df.columns.str.strip().str.replace(" ","_") #cleaning up the column names so they're easier to use (no spaces or weird formatting)
```

```python
df.columns #listing column names to see what I can use
```

```python
df['Percent_Scoring_65_or_Above'].head(10) #checking the first 10 values to see if the data is complete or has bad data
```

```python
df['Percent_Scoring_65_or_Above'] = df['Percent_Scoring_65_or_Above'].replace('s', None) #replacing "s" with NaN (missing value)
```

```python
df['Percent_Scoring_65_or_Above'] = pd.to_numeric(df['Percent_Scoring_65_or_Above'], errors='coerce') #converted the column to numeric values and make any invalid entries to NaN
```

```python
df['Percent_Scoring_65_or_Above'].head(10) #display th first 10 rows to check the cleaned data
```

```python
df['School_Name'].value_counts().head(10) #seeing which schools have the most data
```

```python
school = "Forest Hills High School"
```

```python
school_df = df[df['School_Name'] == school]
```

```python
school_df.shape
```

```python
school_df['Percent_Scoring_65_or_Above'].describe() #getting the stats for my school
```

```python
df['Percent_Scoring_65_or_Above'].describe() #getting the stats for all schools
```

```python
school_avg = school_df['Percent_Scoring_65_or_Above'].mean() #calculating the average of my school
overall_avg = df['Percent_Scoring_65_or_Above'].mean() #calculating the average for all schools

print('Forest Hills High School average:',school_avg) #showing my school's average
print('Overall average:',overall_avg) #showing the overall average

```

```python
import matplotlib.pyplot as plt #importing matplotlib so I can create a chart
```

```python
plt.bar(['Forest Hills High School', 'All Schools'], [school_avg, overall_avg]) # creating a ba chart to compare my school to all schools
plt.title('Average Percent Scoring 65 or Above') # adding a tilte so the chart is clear
plt.ylabel('Percent') #labeling the vertical axis
plt.show() #displaying the chart
```

Conclusion: Forest Hills High School had an average of 72.84, compared to the overall average of 58.07.  Thsi shows the school performs above the overall average.
